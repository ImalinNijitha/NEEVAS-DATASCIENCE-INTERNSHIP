```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Life Expectancy Data.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy')
plt.title('Life Expectancy Over the Years')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print(data.columns)

# Assuming the correct column name is 'Life expectancy ' (note the trailing space if present)
sns.scatterplot(data=data, x='GDP', y='Life expectancy ', hue='Status')
```

```python
plt.figure(figsize=(8, 6))
# Access the column using its actual name from the DataFrame.
# If the name has extra space like 'Life expectancy ', use that instead.
sns.histplot(data['Life expectancy '], kde=True, bins=30)
plt.title('Distribution of Life Expectancy')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy ')  # Assuming 'Life expectancy ' is the correct column name
plt.title('Life Expectancy Over the Years')
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Status', y='Life expectancy ') # Assuming 'Life expectancy ' is the correct column name
plt.title('Average Life Expectancy by Status')
plt.xlabel("Status")
plt.ylabel("Average Life Expectancy")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Status', y='Life expectancy ') # Assuming 'Life expectancy ' is the correct column name
plt.title('Life Expectancy Distribution by Status')
plt.xlabel("Status")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Select only numerical features for correlation analysis
numerical_data = data.select_dtypes(include=['number'])
corr = numerical_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
